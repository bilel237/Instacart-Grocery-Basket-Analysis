#### 5.1: Sampling Data

```python
#Spliting dataframe in two sapmles with the same distribution (split 70/30)
np.random.seed (4)
```

```python
#Create a list holding True/False values to the test np.random.rand()
dev = np.random.rand(len(df))<=0.7
```

```python
# store 70% of the sample in the dataframe big
big = df[dev]
```

```python
# store 30% of the sample in the dataframe small
small =  df[~dev]
```

```python
#verification
len(df)
```

```python
len(big)+ len (small)
```

#### 5.2: Create Line chart

### Step 6: Create a line chart exploring the connections between age and number of dependents

```python
df.head()
```

```python
# defining smaller df using the same 30% subset sample
df_3 = small[['age', 'number_of_dependants']]
```

```python
line_2= sns.lineplot(data = df_3, x = 'age',y = 'number_of_dependants')
```

The line chart shows a fluctuating relationship between customer age and the number of dependents, with no clear trend. This indicates that a one-size-fits-all marketing approach may not be effective. Instead, consider tailoring campaigns to specific age groups and using detailed customer data to personalize marketing strategies, especially for age groups in their 50s and 60s, which show higher variability. This approach will help address the unique family dynamics of each segment, leading to more effective marketing outcomes.

### Step 7: Create a scatterplot of age vs income

```python
# defining smaller df using the same 30% subset sample
df_4 = small[['age', 'income']]
```

```python
#Create a scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = df_4)
```

The scatterplot indicates a wide range of incomes among younger customers (20-40 years) with most concentrated at the lower end, while individuals aged 40 and above show an upward trend in income and higher income brackets, suggesting that median income increases with age. To effectively target these segments, marketing campaigns should offer affordable options to younger customers, premium products to mid-age adults, and high-end products to older customers, utilizing personalized strategies based on detailed customer data to address their unique needs and financial capacities.
