## Step 3 : How do the results for the entire dataframe differ from those of the subset? 

The differences in the results of the entire dataframe compared to the subset are mainly due to the sample size and potential biases in sampling. In the complete dataframe, the averages of orders per department are calculated from a larger amount of data, allowing for a more accurate and representative estimation. On the other hand, the subset, limited to 1,000,000 rows, may not capture the full diversity of purchasing behaviors and might introduce biases, especially if it is not randomly selected. For example, the average order number for Department 10  is close in both subsets (20.197148 in the entire dataframe and 20.091818 in the subset), indicating stable and well-represented ordering patterns. Meanwhile, Department 16  shows a slight variation, with an average of 17.665606 in the entire dataframe and 18.014071 in the subset. This illustrates how subsets can sometimes deviate from the overall trends observed in the complete dataframe, highlighting the importance of using representative samples to obtain accurate results and avoid biased conclusions.

## Step 6 : Create a Spending Flag

```python
# Aggregating Data with transform() Function
ords_prods_merge['avg_prods_price']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Creating a "spending_flag" using loc Function
ords_prods_merge.loc[ords_prods_merge['avg_prods_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_prods_price'] <  10, 'spending_flag'] = 'Low spender'

```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Examining Value Counts on 'spending_flag' for each user
ords_prods_merge.groupby(['spending_flag']).user_id.nunique()
```

## Step 7: Create an Order Frequency Flag

```python
# Aggregating Data with transform() Function
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating flag on order frequency
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'order_frequency_flag'] = 'Non_frequent customer'
ords_prods_merge.loc[(ords_prods_merge['customer_frequency']>10) & (ords_prods_merge['customer_frequency']<=20),'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
# Examining Value Counts on 'order_frequency_flag'
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Examining Value Counts on 'spending_flag' for each user
ords_prods_merge.groupby(['order_frequency_flag']).user_id.nunique()
```
